Описание: Вам необходимо реализовать двусвязный список для управления 
данными клиентов банка. Каждый клиент представляет собой узел в списке и

= Номер счет:
= Баланс счета

Список клиентов должен поддерживать следующие операции°

обновляет баланс клиента по номеру счета

```python
class ClientNode:
    def __init__(self, name, account_number, balance):
        self.name = name
        self.account_number = account_number
        self.balance = balance
        self.prev = None
        self.next = None

class BankClientList():
    def __init__(self) -> None:
        self.head = None
        self.tail = None

    def add_client(self, name, account_number, balance):
        new_client = ClientNode(name, account_number, balance)
        if not self.head:
            self.head = self.tail = new_client #новый клиент становится головой и хвостом списка
        else:
            self.tail.next = new_client
            new_client.prev = self.tail
            self.tail = new_client #новый клиент ствановится хвостом 

    def remove_client(self, account_number):
        current = self.head #Будем начинать поиск с начала списка

        while current and current.account_number != account_number:
            current = current.next

        if not current: #Если дошли до конца списка и клиент не найден
            return False #Возвращаем False - удаление не произошло

        if current.prev: #Если найденный узел не является первым (головой)
            #Перенаправляем указатель next предыдущего узла
            current.prev.next = current.next

        else:
            #Если удаляемый узел - это голва списка, перемещаем голову на следующий узел
            self.head = current.next
        
        if current.next:
            current.next.prev = current.prev
        else:
            self.tail = current.prev

        return True

    def find_client(self, account_number):
        current = self.head #Карент хранит первый элемент из двухсвязного списка
        while current:
            if current.account_number == account_number:
                return True
            current = current.next
        return False


    def list_client(self):
        current = self.head
        while current:
            print(f"Имя {current.name}, номер счета: {current.account_number}, баланс {current.balance}")
            current = current.next #Переход к следующему узлу

    def update_balance(self, account_number, balance):
        current = self.head
        while current:
            if current.account_number == account_number:
                current.balance = balance
            current = current.next

bank_list = BankClientList()
a = "124"
bank_list.add_client("Кристофер", "123", 1000)
bank_list.add_client("Крис", "127", 1000)
bank_list.add_client("Алекср", a, 3000)

bank_list.list_client()
bank_list.remove_client(a)
print(f"Был удален клиент с номером 124")
bank_list.list_client()

   



```

```python
from abc import ABC, abstractmethod

class Weapon(ABC): #интерфейс для реализации оружия
    @abstractmethod
    def use(self) -> str:
        pass

class Sword(Weapon): #Класс оружия - меч
    def use(self) -> str:
        return f"Могучий меч"
    
class Bow(Weapon): #Класс оружия - лук
    def use(self) -> str:
        return f"Лук для стельбы"  

class WeaponCreator(ABC): #Общий класс создателя
    @abstractmethod
    def create_weapon(self) -> Weapon: #Будет создавать конкретный объект
        pass

    def use_weapon(self) -> str:
        weapon = self.create_weapon()
        return f"Создатель: Вы экипировали и использовали {weapon.use()}"
    
class SwordCreator(WeaponCreator): #Класс создателя объекта меча
    def create_weapon(self) -> Weapon:
        return Sword()
    
class BowCreator(WeaponCreator): #Класс создателя объекета лук
    def create_weapon(self) -> Weapon:
        return Bow()
    
def equip_weapon(creator: WeaponCreator) -> None:
    print(f"Игрок: экипирую оружие, не зная его деталей\n {creator.use_weapon()}")

print("Игра: выбираем меч")
equip_weapon(SwordCreator())

print("Игра: выбираем лук")
equip_weapon(BowCreator())
    




```
